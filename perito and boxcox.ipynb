pareto and boxcox

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#creating a pareto dist
x_pr=stats.loggamma.rvs(5,size=5000)+2 #creating a pareto dist
```

```python
x_pr
```

```python
#comparing x_pr with gaussian dist using stats.probplot
stats.probplot(x_pr,plot=plt) #x_pr is not a gaussian distribution
```

```python
#transforming x_pr into a gaussian dist using boxcox
x_prt,_=stats.boxcox(x_pr)
```

```python
print(x_prt,_)
```

```python
stats.probplot(x_prt,dist=stats.norm,plot=plt)
```
