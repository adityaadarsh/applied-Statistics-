### KS TEST

```python
from scipy import stats
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
```

* test_statistics: Dnm
* null hypo: distribution is same

```python
#normal distribution
X=stats.norm.rvs(0,1,size=1000)
sn.set_style('whitegrid')
sn.kdeplot(X,bw=0.5)
plt.show()
```

```python
plt.hist(X, bins=10)
```

```python
stats.kstest(X,'norm')
```

#### accept null hypothesis as Dnm(test statistics) < pvalue

```python
# ranom uniform distribution
Y=np.random.uniform(size=1000)
sn.set_style('whitegrid')
sn.kdeplot(Y,bw=0.5)
plt.show()
```

```python
stats.kstest(Y,'norm')
```

#### reject null hypothesis as Dnm(test statistics) < pvalue
