# Central limit theorem

```python
import numpy as np
import matplotlib.pylab as plt
from random import sample
import seaborn as sn 
from scipy import stats
```

```python
# creating random uniform sample X which does not follow gausian distribution
X=list(np.random.uniform(low=0.0, high=1.0, size=10000))
```

```python
#Random.uniform saples
#using Q-Q plot to check weather Random.uniform saples ~ follows gaussian distribution

stats.probplot(X,dist='norm', fit=True, plot=plt)
plt.show()
```

```python
# using CLT to make 'X' a gaussian distribution
# https://statisticsbyjim.com/basics/central-limit-theorem/  <-- refer this to understand CLT

m=1000 # number of sample
s=[] # for sampling means of samples

for i in range(m):
    s.append(np.mean(sample(population=X, k=1000)))
    
```

```python
# distribution of sampling means of samples
sn.distplot(s)
plt.grid()
plt.show()
```

#### Observation: It looks like X is follwing gausian disribution

### using Q-Q plot to check weather sampling means of samples ~ follows gaussian distribution with mean of mew and std_dev of sigma/root(N)


```python
# using Q-Q plot to check weather sampling means of samples ~ follows gaussian distribution with mean of mew and std_dev of sigma/root(N)
stats.probplot(np.percentile(s,range(0,101)),dist='norm', fit=True, plot=plt)
plt.show()
```
